```python
Nameofmobile=[]
mobileurl=[]
Rating=[]
Review=[]
Offerprice=[]
Orginalprice=[]
Offerpercentage=[]
EMI=[]
Features=[]    
pgno=[]
f=[]

for x in range(1,41):
    start_time = time.time()

    URL='https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=dd7ee1cc-c7f8-48a7-a31e-feefc93b489a&sort=price_desc&page={}'.format(x)
    
    driver = webdriver.Chrome(r'C:\chromedriver_win32\chromedriver.exe')
    driver.get(URL)
    time.sleep(3)
    soup = BeautifulSoup(driver.page_source, 'html5lib')
    driver.close()

    mobile_details=soup.findAll('div',attrs={'class':'_2kHMtA'})
    for mobile in mobile_details:
        
        
        # Name of mobile
        i = mobile.find('div', attrs = {'class':'_4rR01T'})
        if i is None:
            Nameofmobile.append(np.NaN)
        else:
            Nameofmobile.append(i.text)
               
        # rating
        j = mobile.find('div', attrs = {'class':'_3LWZlK'})
        if j is None:
            Rating.append(np.NaN)
        else:
            Rating.append(j.text)

         # reviews
        k = mobile.find('span', attrs = {'class':'_2_R_DZ'})
        if k is None:
            Review.append(np.NaN)
        else:
            Review.append(k.text)


         # Offer price
        l = mobile.find('div', attrs = {'class':'_30jeq3 _1_WHN1'})
        if l is None:
            Offerprice.append(np.NaN)
        else:
            Offerprice.append(l.text)


        #Orginal price
        m = mobile.find('div', attrs = {'class':'_3I9_wc _27UcVY'})
        if m is None:
            Orginalprice.append(np.NaN)
        else:
            Orginalprice.append(m.text)



        # offer percentage

        o = mobile.find('div', attrs = {'class':'_3Ay6Sb'})
        if o is None:
            Offerpercentage.append(np.NaN)
        else:
            Offerpercentage.append(o.text)

          # EMI

        n = mobile.find('div', attrs = {'class':'_2ZdXDB'})
        if n is None:
            EMI.append(np.NaN)
        else:
            EMI.append(n.text)
         # features

        p = mobile.find('ul', attrs = {'class':'_1xgFaf'})
        
        if p is None:
            f.append(np.NaN)
        else:
            f.append(p.text)
        
        
        pgno.append(x)

    end_time = time.time()
    print('Page {} completed in {} seconds'.format(x,end_time-start_time))   


```

### Extracting Data from Features
1. RAM
2. ROM
3. Expanable Memory
4. Screen_cm
5. Screen_inch
6. Rear Camera
7. Front Camera
8. Battery_mAh


```python
mobile_df['Features'][10]
mobile_df['Features'][12]
mobile_df['Features'][14]
mobile_df['Features'][56]
mobile_df['Features'][78]
mobile_df['Features'][79]
mobile_df['Features'][82]
mobile_df['Features'][22]
```

```python
#1.RAM


RAM_regex=re.compile(r'\d+\WGB\WRAM')
RAM_regex.findall(mobile_df.Features[79])

mobile_df['RAM']=mobile_df.Features.apply(lambda x:RAM_regex.findall(x))

mobile_df.RAM=mobile_df.RAM.apply(lambda x:''.join(x))      #Converted list to string
```

```python
#2.ROM


ROM_regex=re.compile(r'[A-Z0-9\s]+ROM')
ROM_regex.findall(mobile_df.Features[0])

mobile_df['ROM']=mobile_df.Features.apply(lambda x:ROM_regex.findall(x))

mobile_df.ROM=mobile_df.ROM.apply(lambda x:''.join(x))      #Converted list to string
```

```python
# 3 Expanable Memory

ExpanableMemory_regex=re.compile(r'Upto.(\d+.[GB|TB].)')
ExpanableMemory_regex.findall(mobile_df.Features[79])


mobile_df['ExpanableMemory']=mobile_df.Features.apply(lambda x:ExpanableMemory_regex.findall(x))

mobile_df.ExpanableMemory=mobile_df.ExpanableMemory.apply(lambda x:''.join(x))      #Converted list to string
```

```python
#4. Screen_cm

Screen_cm_regex=re.compile(r'\d+.\d+ cm')
Screen_cm_regex.findall(mobile_df.Features[0])

mobile_df['Screen_cm']=mobile_df.Features.apply(lambda x:Screen_cm_regex.findall(x))

mobile_df.Screen_cm=mobile_df.Screen_cm.apply(lambda x:''.join(x))      #Converted list to string
```

```python
#5 Screen_inch

Screen_inch_regex=re.compile(r'[0-9a-zA-Z\s.]+(?:inch)')
Screen_inch_regex.findall(mobile_df.Features[0])

mobile_df['Screen_inch']=mobile_df.Features.apply(lambda x:Screen_inch_regex.findall(x))

mobile_df.Screen_inch=mobile_df.Screen_inch.apply(lambda x:''.join(x))      #Converted list to string
```

```python
# 6 Rear_Camera


Rear_Camera_regex=re.compile(r'Display(\d.*?\|)')
Rear_Camera_regex.findall(mobile_df.Features[4])    


mobile_df['Rear_Camera']=mobile_df.Features.apply(lambda x:Rear_Camera_regex.findall(x))
mobile_df.Rear_Camera=mobile_df.Rear_Camera.apply(lambda x:''.join(x))      #Converted list to string
```

```python
# 7 Front_Camera


Front_Camera_regex=re.compile(r'[A-Z0-9\s]+Front Camera')
Front_Camera_regex.findall(mobile_df.Features[4])   


mobile_df['Front_Camera']=mobile_df.Features.apply(lambda x:Front_Camera_regex.findall(x))
mobile_df.Front_Camera=mobile_df.Front_Camera.apply(lambda x:''.join(x))      #Converted list to string
```

```python

#8 Battery_mAh

Battery_mAh_regex=re.compile(r'\d*.mAh')
Battery_mAh_regex.findall(mobile_df.Features[79])    


mobile_df['Battery_mAh']=mobile_df.Features.apply(lambda x:Battery_mAh_regex.findall(x))
mobile_df.Battery_mAh=mobile_df.Battery_mAh.apply(lambda x:''.join(x))      #Converted list to string
```
